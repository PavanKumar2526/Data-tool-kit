```python
#13. Scatter plot with NumPy and Matplotlib:

import numpy as np
import matplotlib.pyplot as plt

# Create random arrays
np.random.seed(42)
x = np.random.rand(100)
y = np.random.rand(100)

# Create scatter plot
plt.figure(figsize=(10, 8))
plt.scatter(x, y, color='red', marker='o', label='Data Points')

# Add horizontal and vertical lines
plt.axhline(y=0.5, color='g', linestyle='--', label='y = 0.5')
plt.axvline(x=0.5, color='b', linestyle=':', label='x = 0.5')

# Set labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Advanced Scatter Plot of Random Values')

# Add legend
plt.legend()

plt.show()
```

```python
#14. Time-series plot with Pandas and Matplotlib:

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create sample data
dates = pd.date_range(start='2023-01-01', periods=100)
temp = np.random.rand(100) * 30 + 10
humidity = np.random.rand(100) * 50 + 30

df = pd.DataFrame({'Date': dates, 'Temperature': temp, 'Humidity': humidity})

# Create plot
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot Temperature
ax1.plot(df['Date'], df['Temperature'], color='red', label='Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='red')
ax1.tick_params(axis='y', labelcolor='red')

# Create second y-axis for Humidity
ax2 = ax1.twinx()
ax2.plot(df['Date'], df['Humidity'], color='blue', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# Set title and show legend
plt.title('Temperature and Humidity Over Time')
fig.legend(loc='upper right', bbox_to_anchor=(1,1), bbox_transform=ax1.transAxes)

plt.show()
```

```python
#15. Histogram with PDF overlay:

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate data
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)

# Create histogram
plt.figure(figsize=(10, 6))
counts, bins, _ = plt.hist(data, bins=30, density=True, alpha=0.7, color='skyblue', label='Histogram')

# Calculate PDF
x = np.linspace(min(data), max(data), 100)
pdf = norm.pdf(x, loc=np.mean(data), scale=np.std(data))

# Plot PDF
plt.plot(x, pdf, 'r-', lw=2, label='PDF')

# Set labels and title
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')
plt.legend()

plt.show()
```

```python
#17. Seaborn quadrant-wise scatter plot:

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(42)
x = np.random.randn(100)
y = np.random.randn(100)

# Create quadrant labels
quadrants = ['Q1' if x[i]>=0 and y[i]>=0 else 'Q2' if x[i]<0 and y[i]>=0 else 'Q3' if x[i]<0 and y[i]<0 else 'Q4' for i in range(len(x))]

# Create scatter plot
plt.figure(figsize=(10, 8))
sns.scatterplot(x=x, y=y, hue=quadrants, palette='deep')

plt.axhline(y=0, color='k', linestyle='--')
plt.axvline(x=0, color='k', linestyle='--')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')
plt.legend(title='Quadrants')

plt.show()
```

```python
#18. Bokeh sine wave plot:

import numpy as np
from bokeh.plotting import figure, show
from bokeh.layouts import column

# Generate data
x = np.linspace(0, 4*np.pi, 100)
y = np.sin(x)

# Create plot
p = figure(title="Sine Wave Function", x_axis_label='X', y_axis_label='Y')

# Add line
p.line(x, y, line_width=2)

# Add grid
p.grid.grid_line_alpha = 0.3

# Show plot
show(column(p))
```

```python
#19. Bokeh random categorical bar chart:

import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, HoverTool

# Generate data
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 100, 5)

# Create ColumnDataSource
source = ColumnDataSource(data=dict(categories=categories, values=values, color=['#'+''.join([np.random.choice(list('0123456789ABCDEF')) for j in range(6)]) for i in range(5)]))

# Create plot
p = figure(x_range=categories, title="Random Categorical Bar Chart", toolbar_location=None, tools="")

# Add bars
p.vbar(x='categories', top='values', width=0.9, source=source, line_color='white', fill_color='color')

# Customize plot
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.xaxis.axis_label = 'Categories'
p.yaxis.axis_label = 'Values'

# Add hover tool
hover = HoverTool(tooltips=[('Category', '@categories'), ('Value', '@values')])
p.add_tools(hover)

# Show plot
show(p)
```

```python
#20. Plotly simple line plot:

import plotly.graph_objects as go
import numpy as np

# Generate data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create figure
fig = go.Figure(data=go.Scatter(x=x, y=y, mode='lines'))

# Update layout
fig.update_layout(
    title='Simple Line Plot',
    xaxis_title='X-axis',
    yaxis_title='Y-axis'
)

# Show plot
fig.show()
```
